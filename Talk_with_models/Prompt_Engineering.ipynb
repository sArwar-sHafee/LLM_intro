```python
import os
import google.generativeai as genai
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 64,
  "max_output_tokens": 8192,
  "response_mime_type": "text/plain",
}

model = genai.GenerativeModel(
  model_name="gemini-1.5-flash",
  generation_config=generation_config,
   system_instruction="You are a helpful assistant",
)

chat_session = model.start_chat(
  history=[
  ]
)

response = chat_session.send_message("Hello Gemini, I am shafee")

print(response.text)
```

## Qwen

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2-0.5B-Instruct")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2-0.5B-Instruct")
```

```python
input_text = "Write a short story about spiderman"
input_ids = tokenizer.encode(input_text, return_tensors="pt")

output = model.generate(input_ids, max_length=256)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("-"*30 + "\n"+ generated_text + "\n" + "-"*30 )
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-2b-it")
model = AutoModelForCausalLM.from_pretrained("google/gemma-2-2b-it", device_map="cuda")
```

```python
input_text = "Write a short story about spiderman"
input_ids = tokenizer.encode(input_text, return_tensors="pt").to("cuda")

output = model.generate(input_ids, max_length=100)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```
